## Import Package

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## Membaca Dataset

```python
df = pd.read_csv('vgsales.csv')

df.head()
```

## Grafik Batang

```python
plt.figure(figsize=(12,7))
plt.bar(df['Genre'], df['Global_Sales'], color='red', width=0.5)
plt.title("Total Pendapatan Genre Game di Global")
plt.xlabel("Genre")
plt.ylabel("Pendapatan di Global")
plt.show()
```

## Kesimpulan
Berdasarkan Grafik diatas pada penjualan Global, Genre Sport memiliki total pendapatan yang paling besar

## Pie Chart

```python
labels = df['Genre'].value_counts().rename_axis('Genre').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(8,8), dpi=200)
plt.title('Genre Game')
plt.pie(df['Genre'].value_counts(), labels=labels['Genre'], autopct='%.1f%%', startangle=90)
plt.show()
```

## Kesimpulan

Dari grafik diatas genre game yang paling banyak diterapkan yaitu Genre Action \
Sedangkan yang paling sedikit diterapkan yaitu Genre Puzzle

## Heatmap

```python
sns.heatmap(df[["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].corr(), cmap='Blues')
plt.title('Heatmap Penjualan Game')

plt.show()
```
